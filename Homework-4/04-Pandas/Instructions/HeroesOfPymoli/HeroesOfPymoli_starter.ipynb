## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# number of unique items
unique = purchase_data["Item Name"].unique()
len(unique)
```

```python
# average purchase price
avg_price = purchase_data["Price"].mean()
print(avg_price)
```

```python
# number of purchases
ID = purchase_data["Purchase ID"].unique()
len(ID)
```

```python
# total revenue
rev = purchase_data["Price"].sum()
print(rev)
```

```python
# summary dataframe
summary = pd.DataFrame({"Unique Items":[len(unique)], "Average Purchase Price":[avg_price],
                                        "Total Number of Purchases":[len(ID)],"Total Revenue":[rev]})
summary["Unique Items"] = summary["Unique Items"].map("{} Items".format)
summary["Average Purchase Price"] = summary["Average Purchase Price"].map("${:.2f}".format)
summary["Total Number of Purchases"] = summary["Total Number of Purchases"].map("{} Purchases".format)
summary["Total Revenue"] = summary["Total Revenue"].map("${:,}".format)

summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# created requested variables and group by gender
gender = purchase_data.groupby(["Gender"])
gender_price = gender["Price"].mean()
gender_count = gender["Purchase ID"].count()
gender_total = gender["Price"].sum()
# only show unique SN's
uniqueusers = purchase_data.drop_duplicates(subset="SN",keep="first")
# group unique users by gender
groupsn = uniqueusers.groupby(["Gender"])
# get purchase per person by dividing total price (grouped by gender) / unique users (grouped by gender)
purchase_per_person = gender_total/groupsn["SN"].count()

# compile into dataframe
summary_data2=pd.DataFrame({"Purchase Count":gender_count,"Average Purchase Price":gender_price,"Total Purchases":gender_total,"Purchase Per Person":purchase_per_person})

#format
summary_data2["Average Purchase Price"] = summary_data2["Average Purchase Price"].map("${:.2f}".format)
summary_data2["Total Purchases"] = summary_data2["Total Purchases"].map("${:.2f}".format)
summary_data2["Purchase Per Person"] = summary_data2["Purchase Per Person"].map("${:.2f}".format)
summary_data2
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
new_purchase_data = purchase_data.groupby(["Age Summary"])
new_price = new_purchase_data["Price"].mean()
new_count = new_purchase_data["Purchase ID"].count()
new_total = new_purchase_data["Price"].sum()
unique = purchase_data.drop_duplicates(subset="SN",keep="first")
groupsn = unique.groupby(["Age Summary"])
# get purchase per person by dividing total price (grouped by gender) / unique users (grouped by gender)
purchase_per_person = new_total/groupsn["Age Summary"].count()
summary_data3=pd.DataFrame({"Purchase Count":new_count,"Average Purchase Price":new_price,"Total Purchases":new_total,"Purchase Per Person":purchase_per_person})

summary_data3["Average Purchase Price"] = summary_data3["Average Purchase Price"].map("${:.2f}".format)
summary_data3["Total Purchases"] = summary_data3["Total Purchases"].map("${:.2f}".format)
summary_data3["Purchase Per Person"] = summary_data3["Purchase Per Person"].map("${:.2f}".format)

summary_data3
```
