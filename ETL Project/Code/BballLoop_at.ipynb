```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

# Mac Users

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.forbes.com/teams/los-angeles-lakers/'
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

title = soup.find_all('span', class_='profile-row__type-title')

team_type = []

for x in title:
    team_type = x.text
    team_types.append(team_type)
    
team_types
```

```python
teams = ['atlanta-hawks','boston-celtics', 'brooklyn-nets',
         'charlotte-hornets', 'chicago-bulls', 'cleveland-cavaliers',
         'dallas-mavericks', 'denver-nuggets', 'detroit-pistons',
         'golden-state-warriors', 'houston-rockets', 'indiana-pacers', 
         'los-angeles-clippers', 'los-angeles-lakers', 'memphis-grizzlies',
         'miami-heat', 'milwaukee-bucks', 'minnesota-timberwolves', 
         'new-orleans-pelicans','new-york-knicks','oklahoma-city-thunder', 
         'orlando-magic', 'philadelphia-76ers', 'phoenix-suns', 'portland-trail-blazers',
        'sacramento-kings', 'san-antonio-spurs', 'toronto-raptors', 'utah-jazz', 'washington-wizards']

# teams = ['atlanta-hawks','boston-celtics'] 

url = 'https://www.forbes.com/teams/'

team_values_list = []

for team in teams:
    query = f'{url}{team}/'
    browser.visit(query)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    result = soup.find_all('span', class_='profile-row__value-text')
    team_name = f'{team}'
    
    team_name = []
    
    for y in result:
        team_value = y.text
        team_name.append(team_value)
    
    team_values_list.append(team_name)
```

```python
df = pd.DataFrame(team_values_list)
df['Team'] = teams
df = df.set_index('Team')
df.columns = team_types
df
```

```python
df.to_csv("Resources/NBA_Teams_Forbes_Data.csv")
```
